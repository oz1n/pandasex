```python
billboard = pd.read_csv('data/billboard.csv')
```

```python
billboard.head()
```

```python
billboard.melt(id_vars=['year','artist','track','time','date.entered'],
              value_name = 'rank', var_name = 'week' )                      # 将每一首歌曲的数据按每周重新清洗
```

```python
(billboard
     .melt(id_vars=['year','artist','track','time','date.entered'],
           value_name = 'rank', 
           var_name = 'week' )
     .groupby('artist')['rank']
     .mean()
)
```

```python
billboard\
     .melt(id_vars=['year','artist','track','time','date.entered'],
           value_name = 'rank', 
           var_name = 'week' )\
     .groupby('artist')['rank']\
     .mean()
```

```python
ebola = pd.read_csv('data/country_timeseries.csv')
```

```python
ebola.head()
```

```python
ebola_long = ebola.melt(id_vars=['Date', 'Day'],
           var_name='cd_country',
           value_name='count')
```

```python
ebola_long.head()
```

```python
ebola_split = ebola_long['cd_country'].str.split('_', expand=True)
```

```python
ebola_long[['status','country']] = ebola_split
```

```python
ebola_long.head()
```

```python
weather = pd.read_csv('data/weather.csv')
```

```python
weather.head()
```

```python
weather_long = weather.melt(id_vars = ['id', 'year', 'month', 'element'],
                           var_name = 'day',
                           value_name = 'temp')
```

```python
weather_long.head()
```

```python
weather_long.pivot_table(index = ['id', 'year', 'month', 'day'],
                        columns='element',
                        values='temp').reset_index()   #创建数据透视表
```

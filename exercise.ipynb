```python
x = np.arange(10)
y = 2*x+2
```

```python
fig, ax = plt.subplots()
ax.plot(x, marker='o', label='x')
ax.plot(y, marker='^', label='y')
ax.legend()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2)
ax1.plot(x, marker='o', label='x')
ax2.plot(y, marker='^', label='y', color='hotpink')
ax1.legend()
ax2.legend()
```

#### Use Shorthand Notation
##### fmt = [color][marker][line]

```python
fig, ax = plt.subplots()
ax.plot(x, 'ro-',label='x')
ax.legend()
ax.set_title('ShortHand Notation')
```

```python
labels = ['a', 'b', 'c']
values = [1, 3, 2]
```

```python
fig, ax = plt.subplots()
bar = ax.bar(labels, values)
patterns = ['/', 'o', '*']
for b in bar:
    b.set_hatch(patterns.pop(0))

#bar[0].set_hatch('/')
#bar[1].set_hatch('o')
#bar[2].set_hatch('*')
```

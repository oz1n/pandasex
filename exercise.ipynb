```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
_ = ax.plot([1,2,3],[4,5,6], label = 'x', marker='o')
_ = ax.plot([1,3,4],[5,6,5], label = 'y', marker='^')
ax.legend()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2)
_ = ax1.plot([1,2,3],[4,5,6], label = 'x', marker='o')
_ = ax2.plot([1,3,4],[5,6,5], label = 'y', marker='^')
ax1.legend()
ax2.legend()
```

```python
x = np.arange(10)
y = 2*x+2
```

```python
fig, ax = plt.subplots()
ax.plot(x, marker='o', label='x')
ax.plot(y, marker='^', label='y')
ax.legend()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2)
ax1.plot(x, marker='o', label='x')
ax2.plot(y, marker='^', label='y', color='hotpink')
ax1.legend()
ax2.legend()
```

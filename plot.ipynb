```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.tip.plot(kind='hist')  #集成在里面
plt.show()
```

```python
cts = tips.smoker.value_counts()
```

```python
cts.plot(kind='bar')
```

```python
sns.countplot(x = "smoker", data = tips)
```

```python
sns.distplot(tips.total_bill)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='smoker')
```

```python
facet = sns.FacetGrid(tips, col="time", row="smoker", hue="sex")
facet.map(plt.scatter, 'total_bill', 'tip')
```

```python
fig, (ax1,ax2) = plt.subplots(1, 2)
ax1.scatter(tips.total_bill, tips.tip)
ax2.hist(tips.total_bill)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
sns.distplot(tips.tip, ax = ax1)
sns.regplot(x=tips.total_bill, y=tips.tip, ax=ax2)
```
